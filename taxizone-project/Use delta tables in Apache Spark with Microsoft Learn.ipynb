```python
df = spark.read.format("csv").option("header","true").load("abfss://MicrosoftLearn@onelake.dfs.fabric.microsoft.com/MicrosoftLearn_LH.Lakehouse/Files/products.csv")

# df now is a Spark DataFrame containing CSV data from "abfss://MicrosoftLearn@onelake.dfs.fabric.microsoft.com/MicrosoftLearn_LH.Lakehouse/Files/products.csv".
display(df.head(10))
```

# Create a managed table

```python
# Managed tables are tables for which both the schema metadata and the data files are managed by Fabric.

df.write.format("delta").saveAsTable("ManagedTable_products")
```

# Create an external table

```python
# create external tables for which the schema metadata is defined in the metastore for the lakehouse,
# but the data files are stored in an external location.

df.write.format("delta").saveAsTable("external_products", path="Files/Products")
```

# Compare managed and external tables

```python
%%sql

DESCRIBE FORMATTED ManagedTable_products;
```

```python
%%sql

DESCRIBE FORMATTED external_products;
```
