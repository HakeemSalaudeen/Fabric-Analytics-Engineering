## **Data cleaning and transformation**

```python
## filter out the rows less than 0

from pyspark.sql.functions import *

# count the rows 
df_count_before_cleaning = df.count()

# Remove invalid records
df = df.filter((col("trip_distance") > 0) & (col("fare_amount") > 0))
df_count_after_cleaning = df.count()

display(df_count_before_cleaning)
display(df_count_after_cleaning)

number_of_deleted_records = df_count_before_cleaning - df_count_after_cleaning

print(f"Removed {number_of_deleted_records} records")
```

```python
## filter out the rows less than 0

from pyspark.sql.functions import *

dff.filter((col("trip_distance") > 0) & (col("fare_amount") > 0)).show()
```

```python
## show the rows less than 0

from pyspark.sql.functions import *

dff.filter((col("trip_distance") < 0) | (col("fare_amount") < 0)).show()
```

```python
# # Cleanse data
# This line essentially converts the "store_and_fwd_flag" column from string values ("Y" or potentially other values) to Boolean values (True or False).
# This ensures that the "lpep_pickup_datetime" and "lpep_dropoff_datetime" columns are stored in a proper timestamp format within PySpark.

df = df.withColumn("store_and_fwd_flag", when(col("store_and_fwd_flag") == "Y", True).otherwise(False))
df = df.withColumn("lpep_pickup_datetime", col("lpep_pickup_datetime").cast("timestamp"))
df = df.withColumn("lpep_dropoff_datetime", col("lpep_dropoff_datetime").cast("timestamp"))

# Display cleansed data to destination
display(df)
```

```python
# print schema to confirm timestamp 

df.printSchema()
```

```python
# Write cleansed data to destination

df.write.format("delta").mode("overwrite").saveAsTable("green_tripdata_2022_08_cleansed")
```

 ## **Exploratory data analysis (EDA)**

```python
# Load data from source
df = spark.read.load("Tables/green_tripdata_2022_09_cleansed", header=True, inferSchema=True)

# Count the number of rows 
df.count()
```

```python
# Group the data by 'VendorID' and count the number of rows in each group. 

from pyspark.sql.functions import *

df.groupBy(col("vendorID")).count().show()

#or 

spark.sql("SELECT vendorID, COUNT (*) AS cnt FROM MicrosoftLearn_LH.green_tripdata_2022_09_cleansed Group by vendorID ORDER BY cnt DESC").show()
```

```python
# Retrieve information about the earliest and latest pickup dates in the dataset.

from pyspark.sql.functions import min, max

oldest_day = df.select(min("lpep_pickup_datetime")).collect()[0][0]
latest_day = df.select(max("lpep_pickup_datetime")).collect()[0][0]

print("Oldest pickup date: ", oldest_day)
print("Latest pickup date: ", latest_day)
```

```python
# Retrieve information about the earliest and latest drop off dates in the dataset.

from pyspark.sql.functions import min, max

oldest_day = df.select(min("lpep_dropoff_datetime")).collect()[0][0]
latest_day = df.select(max("lpep_dropoff_datetime")).collect()[0][0]

print("Oldest dropoff date: ", oldest_day)
print("Latest dropoff date: ", latest_day)
```
