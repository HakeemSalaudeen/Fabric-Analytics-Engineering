```python
df.withColumn
```

```python
#  To create new columns 
# dayofmonth function is used to extract the DOM from the parseddate column 

df.withColumn("dayofmonth",dayofmonth(col("parseddate"))).withColumn("dayofweek",dayofweek(col("parseddate"))).show()
```

```python
#  To create new columns 
# dayofmonth function is used to extract the DOM from the parseddate column 
#  withColumns is better than withColumn 

df.withColumns({"dayofmonth": dayofmonth(col("parseddate")), \
                "dayofweek": dayofweek(col("parseddate")), \
                "dayofyear": dayofyear(col("parseddate"))}).show()
```

```python
# to remane columns

df.withColumnRenamed("account_id","acctid").withColumnRenamed ("frequency","freq").show()
```

```python
# to filter the freqeuncy column with Monthly Issuance

df.where(col("frequency")=="Monthly Issuance").show()
```

```python
# to select two acct with the filter 

df.select(col("account_id"),col("frequency")).where(col("frequency")=="Monthly Issuance").show()
```

```python
# to filter the freqeuncy column with Monthly Issuance

df.filter(col("frequency")=="Monthly Issuance").show()
```

```python
# to filter the freqeuncy column with Monthly Issuance

df.select(col("account_id"),col("frequency")).filter(col("frequency")=="Monthly Issuance").show()
```

```python
#drop null values

df.na.drop().show()
```

```python
#drop null values iif 'ALL' the rows contain null values 

df.na.drop(how = 'all').show()
```
